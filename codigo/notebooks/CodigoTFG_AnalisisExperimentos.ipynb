###  normalizacion de los datos, rango de los valores seas coherente. asi la comparacion sera mas acertada. valores similares y coherentes entre si




```python
experimento_4_1 = experimento_4.replace({'Left': 1})
experimento_4_2 = experimento_4_1.replace({'Right': 2})
experimento_4_3 = experimento_4_2.replace({'Up': 3})
experimento_4_4 = experimento_4_3.replace({'Down': 4})
experimento_4 = experimento_4_4.replace({'Nothing': 0})
```

```python
Simple_experimento_4 = experimento_4
#Escalado simple
Simple_experimento_4['Attention'] = Simple_experimento_4['Attention']/Simple_experimento_4['Attention'].max()
```

```python
Simple_experimento_4['Attention']
```

```python
MinMax_experimento_4 = experimento_4
#Escalado min-max
MinMax_experimento_4['Attention'] = (MinMax_experimento_4['Attention']-MinMax_experimento_4['Attention'].min())/(MinMax_experimento_4['Attention'].max()-MinMax_experimento_4['Attention'].min())
```

```python
MinMax_experimento_4['Attention']
```

```python
ZScore_experimento_4 = experimento_4
#Escalado Z-score
ZScore_experimento_4['Attention'] = (ZScore_experimento_4['Attention']-ZScore_experimento_4['Attention'].mean())/ZScore_experimento_4['Attention'].std()
```

```python
ZScore_experimento_4['Attention']
```
