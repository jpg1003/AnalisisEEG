```python
index = 5

for nombre, exp in datosEEG_all_nombres.items():
    
   

    matriz_correlacion = exp.corr().abs()
    suma_correlaciones = matriz_correlacion.sum()

    componentes_principales = suma_correlaciones.nlargest(index).index
    componentes_peores = suma_correlaciones.nsmallest(index).index
    
    print(nombre)
    print("Las columnas con la mejor correlacion son:", componentes_principales)
```
