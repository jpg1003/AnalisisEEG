```python
def dividir_datos(dataframe):
    
    y = dataframe[TARGET].values
    X = dataframe.drop(columns=[TARGET]).values
    
    # Se dividen los datos. X_train_val sera el 90% de los datos y X_test el 10%
    X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, random_state=SEED, stratify=y, train_size=TRAIN)

    # Se dividen los datos. X_train sera el 90% de los datos de X_train_val y X_val el restante 10%
    X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, random_state=SEED, stratify=y_train_val, train_size=TRAIN)
   
    return X_train, y_train, X_val, y_val, X_test, y_test

# Ejemplo
# X_train, y_train, X_val, y_val, X_test, y_test = dividir_datosp(dataframe)
```

```python
class PlotLearning(Callback):

    def __init__(self, showLR=False):
      self.showLR = showLR              

    def on_train_begin(self, logs={}):
        self.metrics = {}
        for metric in logs:
            self.metrics[metric] = []
            

    def on_epoch_end(self, epoch, logs={}):
        for metric in logs:
            if metric in self.metrics:
                self.metrics[metric].append(logs.get(metric))
            else:
                self.metrics[metric] = [logs.get(metric)]
        
        metric = [x for x in logs if ('val' not in x) and ('lr' not in x)]
        if self.showLR:
          metric.append('lr')
        
        f, axs = plt.subplots(1,len(metric),figsize=(10,4))
        clear_output(wait=True)
        for i,ax in enumerate(axs):
          ax.plot(range(1, epoch + 2), self.metrics[metric[i]],'o--',label=metric[i])
          try:
            ax.plot(range(1, epoch + 2), self.metrics['val_' + metric[i]], 'o--', label='val_' + metric[i])
          except:
            pass
          ax.set_xlabel('# epochs')
          ax.set_ylabel(metric[i])
          ax.legend()
          ax.grid()

        plt.tight_layout()
        plt.show()
        

# Callback para plotear las m茅tricas durante el entrenamiento neuronal.
```

```python
#Crea modelo MLP


def crear_modelo_MLP(timesteps, features, num_classes):
    model_mlp = Sequential([
        Dense(units=UNITS),  
        Dropout(DROPOUT),
        Dense(units=UNITS),  
        Dropout(DROPOUT),
        Dense(num_classes, activation=ACTIVATION)  
    ])
    
    model_mlp.compile(optimizer= OPTIMIZER, metrics=[METRICS], loss=LOSS)
    return model_mlp
```

```python
#Crea modelo RNN

def crear_modelo_RNN(timesteps, features, num_classes):
    model_rnn = Sequential([
        SimpleRNN(units=UNITS, input_shape=(timesteps, features), return_sequences=True),
        Dropout(DROPOUT),
        SimpleRNN(units=UNITS),
        Dropout(DROPOUT),
        Dense(num_classes, activation=ACTIVATION)
    ])
    model_rnn.compile(optimizer= OPTIMIZER, metrics=[METRICS], loss=LOSS)
    return model_rnn
```

```python
#Crea modelo LSTM

def crear_modelo_LSTM(timesteps, features, num_classes):
    model_lstm = Sequential([
        LSTM(units=UNITS, input_shape=(timesteps, features), return_sequences=True),
        Dropout(DROPOUT),
        LSTM(units=UNITS),
        Dropout(DROPOUT),
        Dense(num_classes, activation=ACTIVATION)
    ])
    model_lstm.compile(optimizer= OPTIMIZER, metrics=[METRICS], loss=LOSS)
    return model_lstm

```

```python
#divide el conjunto de datos en ventanas agrupadas por el target y siguiendo la temporalidad de los datos

def preparar_ventanas(data, target, window_size=WINDOWS_SIZE, overlap=OVERLAP):
    if window_size <= overlap:
        raise "OVERLAP no puede ser mayor o igual que el tama帽o de la ventana."
    targets = target.unique()
    # se almacenan los valores unico de target.

    X = [] # Lista vac铆a para almacenar las ventanas de datos
    y = [] # Lista vac铆a para almacenar los targets objetivo de cada venta
    muestras = 0
    for label in targets:
        subdata = data[target == label].values
        subtarget = target[target == label].values
        
        # con un bucle se itera sobre el subconjunto de datos para crear las ventanas
        for i in range(0, len(subdata) - window_size + 1, window_size - overlap):
            # Se crea la ventana de datos para el target
            X.append(subdata[i:i+window_size])
            # Se obtienen los datos correspondientes con el target
            window_targets = subtarget[i:i+window_size].astype(int)
            #se selecciona el target de la venta con el ultimo target de los datos y se a帽ade a la lista
            y_true = window_targets[-1]
            y.append(y_true)

    return np.array(X), np.array(y)
```

```python
# impresion de graficas

def print_chart(ax, titulo_grafica, x_eje, y_eje, datos_lineas, titulos_lineas):
    """
    Visualizaci贸n de la p茅rdida en un subgr谩fico espec铆fico
    ax - ejes del subgr谩fico
    titulo_grafica - t铆tulo de la gr谩fica
    x_eje - t铆tulo del eje X
    y_eje - t铆tulo del eje Y
    datos_lineas - lista de series de valores
    titulos_lineas - t铆tulo de cada linea
    """
    for linea in datos_lineas:
        ax.plot(linea)
    ax.set_title(titulo_grafica)
    ax.set_xlabel(x_eje)
    ax.set_ylabel(y_eje)
    ax.legend(titulos_lineas, loc='best')
```

```python
# en la impresion del dataframe se resalta los tres mejores por colores

def colorear_mayores_3(fila):
    mayores_indices = fila.nlargest(3).index 
    
    primer_mayor = mayores_indices[0]
    segundo_mayor = mayores_indices[1]
    tercer_mayor = mayores_indices[2]
    
    estilos = []
    for columna in fila.index:

        if columna == primer_mayor:
            estilo = 'background-color: green'
        elif columna == segundo_mayor:
            estilo = 'background-color: yellow'
        elif columna == tercer_mayor:
            estilo = 'background-color: orange'
        else:
            estilo = ''
        estilos.append(estilo)
    
    return estilos
```
