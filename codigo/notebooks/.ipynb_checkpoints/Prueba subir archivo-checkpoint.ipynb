```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import io

# Widget para la carga del archivo
upload = widgets.FileUpload(accept='.csv', multiple=False)

# Variable global para almacenar el DataFrame
dafaframe_total = pd.DataFrame()

def on_upload_change(change):
    global dafaframe_total
    # Verificar si hay archivos cargados
    if upload.value:
         # Obtener el contenido del archivo cargado
        uploaded_file = upload.value[0]  # Acceder al primer elemento de la tupla
        content = uploaded_file['content']
        
        # Convertir el contenido de memoryview a bytes y despues a string
        content_bytes = content.tobytes()
        
        content_str = content_bytes.decode('utf-8')
        dafaframe_total = pd.read_csv(io.StringIO(content_str), delimiter=';')
        
    else:
        print("No se ha cargado ningún archivo")

# Observador del cambio de archivo cargado
upload.observe(on_upload_change, names='value')

# Mostrar el widget
display(upload)


```
