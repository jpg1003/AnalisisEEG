```python
display(HTML(html_h4('Realizo la importacion a Dataframe de los datos de los csv creados en la etapa de Prepocessing: ','purple')))
experimento_1 = pd.read_csv('../datos/experimento_1.csv', sep=',')
experimento_2 = pd.read_csv('../datos/experimento_2.csv', sep=',')
experimento_3 = pd.read_csv('../datos/experimento_3.csv', sep=',')
experimento_4 = pd.read_csv('../datos/experimento_4.csv', sep=',')
exp_all = pd.read_csv('../datos/experimento_all.csv', sep=',')

experimentos = [experimento_1, experimento_2, experimento_3,experimento_4]
experimentos_name = {'experimento_1':experimento_1, 'experimento_2':experimento_2, 'experimento_3':experimento_3,'experimento_4':experimento_4}
```

### Correlaciones ###

```python
exp_all.corr()
```

```python
sns.heatmap(exp_all.corr(), cmap="BrBG") 
```

```python
index = 5

for nombre, exp in experimentos_name.items():
    
   

    matriz_correlacion = exp.corr().abs()
    suma_correlaciones = matriz_correlacion.sum()

    componentes_principales = suma_correlaciones.nlargest(index).index
    componentes_peores = suma_correlaciones.nsmallest(index).index
    
    print(nombre)
    print("Las columnas con la mejor correlacion son:", componentes_principales)
```

```python

matriz_correlacion = exp_all.corr().abs()
suma_correlaciones = matriz_correlacion.sum()

componentes_principales = suma_correlaciones.nlargest(index).index
componentes_peores = suma_correlaciones.nsmallest(index).index

print("experimento_all")
print("Las columnas con la mejor correlacion son:", componentes_principales)
```
