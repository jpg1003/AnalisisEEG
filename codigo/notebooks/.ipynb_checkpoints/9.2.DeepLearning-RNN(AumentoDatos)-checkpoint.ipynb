```python
datosEEG_seg1_standarizado = pd.read_csv('../datos/csv/seg1_stscaler.csv', sep=',')
datosEEG_std_after_resampled = pd.read_csv('../datos/csv/datosEEG_std_after_resampled.csv', sep=',')
datosEEG_std_before_resampled = pd.read_csv('../datos/csv/datosEEG_std_before_resampled.csv', sep=',')
datosEEG_std_tres_segmentos = pd.read_csv('../datos/csv/datosEEG_std_tres_segmentos.csv', sep=',')

datosEEG_all = [datosEEG_std_after_resampled, datosEEG_std_before_resampled]

nombre_clases = valores_key(datosEEG_seg1_standarizado)
```

# RNN (Recurrent Neural Networks)

```python
model_rnn_rs = crear_modelo_RNN(TIMESTEPS, FEATURES, NUM_CLASES)

display(HTML(html_h5('Entreno los conjuntos de datos para el  modelo RNN con aumento de datos: ', 'purple')))
display(HTML(html_h5('Se imprimen a tiempo real las ejecuciones de cada conjunto de datos: ', 'blue')))
```

```python
accuracy_score_rnn_rs = []
accuracy_score_test_rnn_rs =[]
matriz_confusion_rnn_rs = []
matriz_confusion_test_rnn_rs = []
total_rnn_rs_trained = []

for dataset in datosEEG_all:
    
    X_train, y_train, X_val, y_val, X_test, y_test = dividir_datos(dataset)
    
    X_train, X_val, X_test = reshapeo(X_train, X_val, X_test)
            
    # Entrenar el modelo
    model_rnn_trained_rs = model_rnn_rs.fit(X_train, y_train, epochs=EPOCHS, batch_size=BACH_SIZE, validation_data= (X_val, y_val), callbacks=[early_stopping, REDUCE_RL, model_checkpoint, plot_learning])
    
    y_pred = model_rnn_rs.predict(X_val)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    y_pred_test = model_rnn_rs.predict(X_test)
    y_pred_classes_test = np.argmax(y_pred_test, axis=1)

    accuracy_score_rnn_rs.append(accuracy_score(y_val, y_pred_classes))
    matriz_confusion_rnn_rs.append(confusion_matrix(y_val, y_pred_classes))
    
    accuracy_score_test_rnn_rs.append(accuracy_score(y_test, y_pred_classes_test))
    matriz_confusion_test_rnn_rs.append(confusion_matrix(y_test, y_pred_classes_test))

    total_rnn_rs_trained.append(model_rnn_trained_rs)
```

```python

X_train, y_train, X_val, y_val, X_test, y_test = dividir_datos_deep_sr(datosEEG_std_tres_segmentos, datosEEG_seg1_standarizado, TARGET)

X_train, X_val, X_test = reshapeo(X_train, X_val, X_test)

# Entrenar el modelo
model_rnn_rs_trained1 = model_rnn_rs.fit(X_train, y_train, epochs=EPOCHS, batch_size=BACH_SIZE, validation_data= (X_val, y_val), callbacks=[early_stopping, REDUCE_RL, model_checkpoint, plot_learning])

y_pred = model_rnn_rs.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)

y_pred_test = model_rnn_rs.predict(X_test)
y_pred_classes_test = np.argmax(y_pred_test, axis=1)

accuracy_score_rnn_rs.append(accuracy_score(y_val, y_pred_classes))
matriz_confusion_rnn_rs.append(confusion_matrix(y_val, y_pred_classes))

accuracy_score_test_rnn_rs.append(accuracy_score(y_test, y_pred_classes_test))
matriz_confusion_test_rnn_rs.append(confusion_matrix(y_test, y_pred_classes_test))

total_rnn_rs_trained.append(model_rnn_rs_trained1)
```

```python
display(HTML(html_h5('Estos son los resultados para Tasa de acierto y su matrices de confusin para el subconjunto de datos validacin: ', 'blue')))
```

```python

datos_rnn_rs = pd.DataFrame(accuracy_score_rnn_rs, columns=['RNN_RS_VAL'], index=(titulos_all_rs))

display(datos_rnn_rs)

mc_rnn = []

for matrices_rnn in matriz_confusion_rnn_rs:

    mc_rnn.append(pd.DataFrame(matrices_rnn, index=nombre_clases, columns=nombre_clases))

#https://scikit-learn.org/stable/modules/generated/sklearn.metrics.ConfusionMatrixDisplay.html

# Creo una figura y ejes para dividir el espacio en cuatro subgrficos
fig, axs = plt.subplots(1, 3, figsize=(10, 12))

# Matrices de confusin y ttulos para cada subgrfico
dataframes = [mc_rnn[0], mc_rnn[1], mc_rnn[2]]


# Itero sobre los DataFrames y ttulos para imprimir matriz de confusin en un subgrfico
for i, (dataframe, titulo) in enumerate(zip(dataframes, titulos_all_rs)):

    ax = axs[i]
   
    # Se crea la impresion de la matriz de confusin y se implime
    disp = ConfusionMatrixDisplay(confusion_matrix=dataframe.values, display_labels=(nombre_clases))
    disp.plot(ax=ax, colorbar=False, cmap='Blues')
    
    # Aado el ttulo al subgrfico
    ax.set_title(titulo)

# Se ajusta el espaciado entre subgrficos
plt.tight_layout()

# Se imprime el grfico
plt.show()
```

```python

display(HTML(html_h5('Imprimo los graficos Tasa de acierto y loss durante los entrenamientos de los modelos : ', 'blue')))

display(HTML(html_h5('Se imprimern en este orden "All Segmentos after", "All Segmentos before", "Test contra Datos Reales" " : ', 'blue')))

# Crear una figura con 6 subgrficos (3 filas y 2 columnas)
fig, axes = plt.subplots(3, 2, figsize=(10, 15))

# Asegurarse de que haya suficiente espacio entre los subgrficos
plt.tight_layout(pad=5.0)

# Alternar entre 'loss' y 'accuracy' para cada par de subgrficos
for idx, history in enumerate(total_rnn_rs_trained[:4]):
    i, j = divmod(idx * 2, 2)
    # Imprimir la grfica de prdida en el primer subgrfico del par
    print_chart(axes[i, j], "Model Accuracy", "Epoch", "Accuracy", 
            [history.history['accuracy'], history.history['val_accuracy']], ['Train', 'Val'])
    
    # Imprimir la grfica de precisin (accuracy) en el segundo subgrfico del par
    i, j = divmod(idx * 2 + 1, 2)
    print_chart(axes[i, j], "Model Loss", "Epoch", "Loss", 
                [history.history['loss'], history.history['val_loss']], ['Train', 'Val'])

# Mostrar todas las grficas en una sola figura
plt.show()
```
